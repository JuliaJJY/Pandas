### 데이터 병합 방법
1) merge : 기준이 되는 열(key)을 갖고 데이터 프레임 병합
    - df1.merge(df2)
    - pd.merge(df1, df2, on='', how='')
2) concentrate : 연결

````python
#data frame 생성
df1 = pd.DataFrame({'고객명':['춘향','춘향','몽룡'],
    '날짜' : ['2018-01-01','2018-01-02','2018-01-01'],
    '데이터':[20000,30000,100000]})

df2 = pd.DataFrame({'고객명':['춘향','몽룡'],
    '데이터':['여자','남자']})

pd.merge(df1,df2, on ='고객명', how='outer')
````

#### left_on, right_on : merge 기준 열 명시 함수

````python
#키가 되는 기준열이 두 df 에서 서로 다르면 left_on, right_on 인수를 사용해서 기준열을 명시 해야함.
df1=pd.DataFrame({
    '이름' :['영희','철수','철수'],
    '성적' :[90,80,80]})

df2 = pd.DataFrame({
    '성명' :['영희','영희','철수'],
    '성적2':[100,80,90]})
````

````python
pd.merge(df1,df2, left_on='이름',right_on='성명') 
````

#### 인덱스 기준으로 merge 하기
left_index=True 
right_index=True

````python
df1 = pd.DataFrame({
    '도시': ['서울','서울','서울','부산','부산'],
    '연도': [2000,2005,2010,2000,2005],
    '인구':[9853972,9762546,9631482,3655437,3512547]})

df2=pd.DataFrame(
    np.arange(12).reshape((6,2)),
    index=[['부산','부산','서울','서울','서울','서울'],
          [2000,2005,2000,2005,2010,2015]],
    columns=['데이터1','데이터2'])

#reshape : np.reshape(배열, 차원) 혹은 배열.reshape(차원) 으로 나타냄

print(df1)
print(df2)
````

````python
#행, 열 조건으로 데이터 추출
df2.loc['부산'].loc[2000]

# df1.loc['부산'].loc[2000] - 오류 : df1에서 부산은 인덱스가 아니기 때문
````

````python
#인덱스만 보기
df2.index
````

````python
#인덱스 기준으로 merge 하기
pd.merge(df1,df2,left_on=['도시','연도'],right_index = True)    
#df1에서는 일반 열이기 때문에 left_on 사용하여 인덱스가 될 열 이름을 적었지만, df2는 인덱스를 갖고 있기 때문에 right_index =True 사용
````

````python
df1 = pd.DataFrame([[1.,2.],[3.,4.],[5.,6.]],
index=['a','c','e'],columns=['서울','부산'])
print(df1)

df2=pd.DataFrame([[7.,8.],[9.,10.],[11.,12.],[13.,14.]],
    index=['b','c','d','e'],columns=['대구','광주'])
print(df2)
````

````python
#공통 컬럼이 없고, 양쪽 데이터프레임에 key가 모두 인덱스인 경우
# pd.merge(df1,df2,how='outer')  #공통 컬럼 없어서 오류 남
pd.merge(df1,df2,left_index=True, right_index=True, how='outer') # outer 적용 안하면 inner 로 적용 됌
````

### Concat 
- 데이터 연결시 사용
```
pd.concat(objs,         # Series, DataFrame, Panel object   
          axis=0,       # 0: 위+아래로 행결합, 1: 왼쪽+오른쪽으로 열결합  
          join='outer', # 'outer': 합집합(union), 'inner': 교집합(intersection)  
          ignore_index=False,  # **False: 기존 index 유지**, **True: 기존 index 무시**    
          keys=None,    # 계층적 index 사용하려면 keys 튜플 입력)  

````python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'], 
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'E': ['C4', 'C5', 'C6', 'C7'],
                    'F': ['D4', 'D5', 'D6', 'D7']},
                   index=[0, 1, 2, 3])

df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11'],
                    'O': ['D8', 'D9', 'D10', 'D11']},
                   index=[1,2,3,4])
df1
df2
df3
````

````python
#concat() : 위 아래로 단순 병합
result = pd.concat([df1,df2,df3])
result
````

````python
result=pd.concat([df1,df2,df3], keys=['x','y','z'])
result #인덱스가 중복되는 경우 멀티인덱스 생성
````

````python
result.loc['x'].loc[0,'A']
````

````python
#keys= 상위레벨 인덱스 설정
result=pd.concat([df1,df2,df3], join='inner',keys=['x','y','z'])
result #공통적으로 나타나는 열만 표혐
````

````python
#다중 인덱스에서 행 데이터 접근하기 - 연산자를 이용한 체인 인덱싱 (*열 데이터 접근은 밑에)
result.loc['y'].loc[1]
result.loc['y'].loc[1:2]
````

````python
result = pd.concat([df1,df2,df3], join= 'inner')
result
````

````python
#ignore_index = True : 기존 인덱스 제거 후 제로베이스 인덱스 설정
result = pd.concat([df1,df2,df3], join= 'inner',ignore_index=True)
result
````

#### concat 활용 열방향, 행방향 결합

````python
df1=pd.DataFrame(np.arange(6).reshape(3,2),
    index=['a','b','c'],
    columns=['데이터1','데이터2'])
df1
df2=pd.DataFrame(5+np.arange(4).reshape(2,2),
    index=['a','c'],
    columns=['데이터2','데이터4'])
df2
````

````python
pd.concat([df1,df2],axis=1) #열방향 결합, concat은 how  대신 join으로 방식  표현!
````

````python
pd.concat([df1,df2],axis=0)
````

````python
pd.concat([df1,df2],axis=1, join ='inner')
````

````python
#위 방식에서 데이터2만 추출
test=pd.concat([df1,df2],axis=1,join='inner')
test.데이터2
````

````python
#다중 인덱스에서 열 추출하기!!

test=pd.concat([df1,df2],axis=1, join ='inner', keys=['x','y'])
print(test)
print(test.x)
print(test.x.데이터1)
````
